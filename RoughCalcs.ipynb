```python
import yfinance as yf
import pandas as pd

def get_stock_data(symbol, start_date, end_date):
    # Download historical data for the specified date range
    data = yf.download(symbol, start=start_date, end=end_date)

    # Calculate the 20 SMA, 50 EMA, and RSI
    data['SMA_20'] = data['Close'].rolling(window=20).mean()
    data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()

    # Calculate RSI
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    data['RSI'] = 100 - (100 / (1 + rs))

    # Store the result in a new DataFrame
    final_data = data[['Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'SMA_20', 'EMA_50']]

    return final_data

# Example usage:
symbol = 'AAPL'
start_date = '2021-01-01'
end_date = '2024-10-22'

# Call the function and store the result in a DataFrame
final_data = get_stock_data(symbol, start_date, end_date)

# Display the last few rows of the DataFrame
print(final_data.tail())

# Optional: Save to a CSV file
final_data.to_csv(f'{symbol}_technical_indicators.csv')

print(f"Daily data saved to {symbol}_technical_indicators.csv")
```

```python
import requests
import pandas as pd

def fetch_options_data(symbol, date, api_key):
    # AlphaVantage API URL
    url = f"https://www.alphavantage.co/query?function=HISTORICAL_OPTIONS&symbol={symbol}&date={date}&apikey={api_key}&datatype=csv"
    
    # Request the data
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Convert the CSV response into a pandas DataFrame
        from io import StringIO
        data = StringIO(response.text)
        df = pd.read_csv(data)
        return df
    else:
        print(f"Error: Unable to fetch data (status code {response.status_code})")
        return None

def calculate_call_put_volume_premiums(df):
    # Separate call and put options
    call_options = df[df['type'] == 'call']
    put_options = df[df['type'] == 'put']
    
    # Calculate total volume and premium for calls
    total_call_volume = call_options['volume'].astype(float).sum()
    total_call_premium = (call_options['volume'].astype(float) * call_options['mark'].astype(float)).sum()

    # Calculate total volume and premium for puts
    total_put_volume = put_options['volume'].astype(float).sum()
    total_put_premium = (put_options['volume'].astype(float) * put_options['mark'].astype(float)).sum()
    
    # Calculate put/call volume ratio
    put_call_volume_ratio = total_put_volume / total_call_volume if total_call_volume > 0 else None
    
    # Return the results
    return total_call_volume, total_call_premium, total_put_volume, total_put_premium, put_call_volume_ratio

# Set your parameters
symbol = 'TSLA'
date = '2023-09-29'
api_key = '9GZJKCNBTDQJL9BR'  # Replace with your actual API key

# Fetch the options data
options_df = fetch_options_data(symbol, date, api_key)

if options_df is not None:
    # Calculate total call/put volume, premiums, and the put/call volume ratio
    total_call_volume, total_call_premium, total_put_volume, total_put_premium, put_call_volume_ratio = calculate_call_put_volume_premiums(options_df)

    # Display the results
    print(f"Total Call Volume: {total_call_volume}")
    print(f"Total Call Premium: {total_call_premium}")
    print(f"Total Put Volume: {total_put_volume}")
    print(f"Total Put Premium: {total_put_premium}")
    print(f"Put/Call Volume Ratio: {put_call_volume_ratio}")
```

```python
import yfinance as yf
import talib as ta
import pandas as pd

# Download historical data for the selected stock (AAPL)
ticker = 'AAPL'
data = yf.download(ticker, start='2020-01-01', end='2023-01-01')

# Download VIX data
vix_data = yf.download('^VIX', start='2020-01-01', end='2023-01-01')[['Close']].rename(columns={'Close': 'VIX'})

# Merge AAPL data with VIX data based on the Date index
data = data.merge(vix_data, left_index=True, right_index=True, how='left')

# Pivot Point Calculation (used for Support and Resistance)
pivot = (data['High'] + data['Low'] + data['Close']) / 3

# First Resistance and Support
data['R1'] = 2 * pivot - data['Low']
data['S1'] = 2 * pivot - data['High']

# Second Resistance and Support
data['R2'] = pivot + (data['High'] - data['Low'])
data['S2'] = pivot - (data['High'] - data['Low'])

# Calculate technical indicators
data['20EMA'] = ta.EMA(data['Close'].to_numpy().flatten(), timeperiod=20)
data['50SMA'] = ta.SMA(data['Close'].to_numpy().flatten(), timeperiod=50)
data['RSI'] = ta.RSI(data['Close'].to_numpy().flatten(), timeperiod=14)

# Final DataFrame with proper column names including VIX
final_data = data[['Open', 'High', 'Low', 'Close', 'Volume', 'VIX', 'R1', 'R2', 'S1', 'S2', '20EMA', '50SMA', 'RSI']].copy()

# Add ticker column at the beginning
final_data.insert(0, 'Ticker', ticker)

# Reset index to avoid adding the old index as a column
final_data.reset_index(inplace=True)

# Round off values
final_data = final_data.round(2)

# Export to CSV
final_data.to_csv(f"{ticker}_technical_analysis.csv", index=False)

# Show the last few rows
print(final_data.tail())
```

```python
import yfinance as yf
import pandas as pd

def calculate_technical_indicators(df):
    # Calculate the 20-day EMA
    df['20EMA'] = df['Close'].ewm(span=20, adjust=False).mean()

    # Calculate the 50-day SMA
    df['50SMA'] = df['Close'].rolling(window=50).mean()

    # Calculate RSI (Relative Strength Index)
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))

    # Round off to 2 decimals
    df = df.round(2)

    return df

def get_vix_data(start_date="2012-01-01"):
    # Download the VIX data from Yahoo Finance
    vix_data = yf.download('^VIX', start=start_date)
    # Keep only the 'Adj Close' column and rename it to 'VIX'
    vix_data = vix_data[['Adj Close']].rename(columns={'Adj Close': 'VIX'}).round(2)
    return vix_data

def download_stock_data(ticker, start_date="2012-01-01"):
    # Download the stock data
    stock_data = yf.download(ticker, start=start_date)
    
    # Add the company name as a new column
    stock_data['company_name'] = ticker
    
    # Calculate technical indicators
    stock_data = calculate_technical_indicators(stock_data)
    
    # Get the VIX data and merge it based on the Date index
    vix_data = get_vix_data(start_date)
    stock_data = stock_data.merge(vix_data, how='left', left_index=True, right_index=True)
    
    # Reset the index to move the date into a column
    stock_data.reset_index(inplace=True)
    
    # Save the result to a CSV file
    csv_filename = f'{ticker}_data.csv'
    stock_data.to_csv(csv_filename, index=False)
    
    return stock_data

# Example usage with dynamic tickers
tickers = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'META', 'TSLA', 'NVDA']
for ticker in tickers:
    data = download_stock_data(ticker)
    print(f"Data for {ticker} saved successfully.")
```
